Base 2: $f(x) = 2^x$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)
y = 2 ** x

fig, ax = plt.subplots()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(x, y)
```

## exponential function

$f(x) = A  e^{kx}$ where $ A, K \in I$

A > 0, K > 0

```python
A = 1
e = 2
k = 1

x = np.linspace(-3, 3, 100)
y = A * e ** (k * x)

fig, ax = plt.subplots()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(x, y)
```

A < 0, K > 0

```python
A = -3
e = 2
k = 1

x = np.linspace(-3, 3, 100)
y = A * e ** (k * x)

fig, ax = plt.subplots()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(x, y)
```
