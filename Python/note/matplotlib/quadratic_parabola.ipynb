# Quadratic / parabola

$f(x) = ax^2 + bx + c$

example:$f(x) = x^2 + 3x + 6$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20, 20, 100)
y = x ** 2 + 3 * x + 6

plt.plot(x, y)
```

# Cubic

$f(x) = ax^3 + bx^2 + cx + d$

$a, b, c, d \in R$ 

Example: $2x^3 + 3x^2 -2x -3$

```python
x = np.linspace(-20, 20, 100)
y = 2 * x**3 + 3 * x**2 -2*x -3

plt.plot(x, y)
```
