# MatPlotLib Basics

Matplotlib graphs on `Figure`s. Each of which can contain one or more `Axes`(座標系統). 

Easy way to draw is by `pyplot.subplots()`, it returns a tuple of `(Figure, Axis)` objects.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
# ax.plot(x-axis, y-axis)
ax.plot([1, 2, 3, 4], [3, 1, 2, 2])
```

Parts of a Figure
![Parts of a Figure](https://matplotlib.org/stable/_images/anatomy.png)

## Axes

Usually made up of two `Axis` objects, x-axis & y-axis in 2D.

Set label at x by `ax.set_xlabel('x label')`
Set label at y by `ax.set_ylabel('y label')`

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [3, 2, 1, 2])
ax.set_xlabel('this is x')
ax.set_ylabel('this is y')
```

### Axis scales

[X scale](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xscale.html)

[Y scale](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_yscale.html)

```python
fig, ax = plt.subplots()
x = np.linspace(1, 5, 10)
ax.plot(x, x**2)
ax.set_yscale('log')
```

### Axis ticks

```python
fig, ax = plt.subplots()
x = np.linspace(1, 5, 10)
ax.plot(x, x**2)
ax.set_xticks([1, 3, 5])
```

## Coding style

- explictly, OO-style
- implictly, pyplot-style

### OO-Style

Works on the return objects

```python
x = np.linspace(0, 2, 100)  # Sample data.

fig, ax = plt.subplots(figsize=(5, 2.7))

# plots 3 line
ax.plot(x, x, label='linear')  
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic') 

ax.set_xlabel('x label') 
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()
```

### pyplot style

directly on the pyplot method

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
```

## Styling

### Line style

- color
- line width
- line style

```python
fig, ax = plt.subplots()

# pyplot style
ax.plot(x, x ** 2, label='dd1',color='r', linewidth=10, linestyle='--')

# OO style
l = ax.plot(x, x) # l is a 2D list
l[0].set_label('gg1')
l[0].set_linestyle(':')
l[0].set_color('g')
l[0].set_linewidth(10)
fig.legend()
```

### add grid

```python
x = np.linspace(1, 5, 10)
fig, ax = plt.subplots()
ax.plot(x, x ** 2)
ax.grid('True')
```

## Labelling

```python
fig, ax = plt.subplots()
x = np.linspace(1, 5, 20)
ax.plot(x, x ** 3)

ax.set_xlabel('X_axis')
ax.set_ylabel('Y_axis')

at_x = 3
at_y = 80
ax.text(at_x, at_y, r'hello')
```

## Save it to a File

```python
x = np.linspace(1, 5, 10)
plt.plot(x, x ** 2)
plt.savefig('./MyPlot.png', format='png')
```

## Pie Chart

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0]
labels = ['India', 'United States', 'Russia', 'China', 'Europe']
plt.pie(values, colors= colors, labels=labels, explode = explode)
plt.title('Student Locations')
plt.show()
```

## Bar Chart

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
plt.bar(range(0,5), values, color= colors)
plt.show()
```

## Scatter Plot

Syntax: `plt.scatter(arrayA, arrayB)`

where arrayA and arrayB is same length

```python
from pylab import randn

X = randn(500)
Y = randn(500)
plt.scatter(X,Y)
plt.show()
```

## Histogram

```python
incomes = np.random.normal(27000, 15000, 10000)
plt.hist(incomes, 50)
plt.show()
```

## Box & Whisker Plot

Useful for visualizing the spread & skew of data.

The red line represents the median of the data, and the box represents the bounds of the 1st and 3rd quartiles.

So, half of the data exists within the box.

The dotted-line "whiskers" indicate the range of the data - except for outliers, which are plotted outside the whiskers. Outliers are 1.5X or more the interquartile range.

This example below creates uniformly distributed random numbers between -40 and 60, plus a few outliers above 100 and below -100:

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```
