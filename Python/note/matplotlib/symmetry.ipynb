# Symmetry

## odd symmetry

odd symmetry $f(-x) = - f(x)$

example: $f(x) = 3x + 6$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = 3 * x **3 + 6

fig, ax = plt.subplots()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(x, y)
```

## even symmetry

even symmetry $f(-x) = f(x)$

example: $f(x) = -x^2 + 10$

```python
x = np.linspace(-10, 10, 100)
y = - x ** 2 + 10

fig, ax = plt.subplots()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(x, y)
```
