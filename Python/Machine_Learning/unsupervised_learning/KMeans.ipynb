# K-means

K-means is unsupervised learning, which means you don't have to label your data. When you don't know where to start with a dataset, it's a good start point.

You define a *K* number of cluster. The model will based on the mathematical model to partition your data into K group. It's used to find a hidden pattern.

You define `K=3`, it gives you 3 cluster.

## Demo

Let's take atlantic city salary data for this demo.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('../data/atl_salary_data_2015.csv')

X = df[['age', 'annual.salary']]
X = np.array(X)
```

We use *age and anuual salary*. We then plot it with k cluster model. 

Here we tell the model that we want 5 groups:

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

k = 5

model = KMeans(n_clusters=k)
model = model.fit(X)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(X[:, 0], X[:, 1], c=model.labels_.astype(float))
plt.show
```

Let's use `sklearn.preprocessing.scale` to preprocess the data, and visualize it. Still, 5 partitions.

```python
model = model.fit(scale(X))
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(X[:, 0], X[:, 1], c=model.labels_.astype(float))
plt.show
```

`scale(data)` is used when the data is largely sparse datasets. It uses some math trick to make more data less skewed. More general discussion: [stackoverflow link](https://stackoverflow.com/a/42332663)

For our dataset, this is actually good. For the first graph, it seems that it groups based on salary only. Each partition covers all age. However, when `scale()` in use, we get a different story.

Now, Let's try 3 clusters with `scale()`

```python
k = 3

model = KMeans(n_clusters=k)
model = model.fit(scale(X))
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(X[:, 0], X[:, 1], c=model.labels_.astype(float))
plt.show
```

3 clusters seems underfit. We may stick to 5 cluster with the preprocessing by `sklearn.preprocessing.scale()`!
