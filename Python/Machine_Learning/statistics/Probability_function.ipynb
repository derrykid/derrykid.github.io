# Types of Data distribution

## Probability density function - continuous data

> The probability density function tells the probability of a given range of values occurring for continuous data.

A continuous data is fluctuate over time. A temperature in a day, every second, every minute might be different. Any value falls in +- 1 standard deviation, will have the likelihood of `2*(34.1%)` of happening. That's probability density function

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)  # (start, stop, step)
plt.plot(x, norm.pdf(x))
```

### Uniform distribution

> A uniform distribution has equal probability across any given value within the
range that you define

In uniform distribution, you would see a flat line on the probability distribution function because there is basically a constant probability. 

Every value, every range of values has an equal chance of appearing as any other value.

```python
low = -10
high = 10
size = 100_000
values = np.random.uniform(low, high, size)
plt.hist(values, 50)
plt.show()
```

### Normal or Gaussian distribution

> It's a type of continuous probability distribution for a real-valued random variable.

Normal distribution, aka Gaussian distribution, bell curve, 常態分佈, 正態分佈, etc

```python
import matplotlib.pyplot as plt

mu = -3    #mean
sigma = 3
step = 100_000
x = np.random.normal(mu, sigma, step) 

buckets = 50
plt.hist(x, buckets)
plt.show()
```

### Exponential probability distribution, aka Power Law

> Fall off in an exponential manner

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Probability mass function - for discrete data

> The probabilities of given discrete values occurring in a dataset.

This used to deal with *discrete* data.

### Binomial probability mass function

> Refers to the probability of exactly x successes on n repeated trials in an experiment.

What is the probability of getting 6 heads, when you toss a coin 10 times?

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
plt.show()

# if this graph is a line, probably there's a bug.
# it supposes to be a histogram
```

### Poisson probability mass function

> If you have some information about the average number of things that happen in a given time period, this probability mass function can give you a way to predict the odds of getting another value instead.

A website, on average I get 500 visitors per day. I can use this function to estimate the probability of seeing some other value on a specific day. 

For example, what's the odds of seeing 550 visitors on a given day? That's what a Poisson probability mass function can give you:

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(mu, x))
```
